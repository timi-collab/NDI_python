## Array creation

Useful functions to create arrays from scratch as an alternative to
converting base Python data types.

(stored as floats by default).

```
np.ones((10, 2))
```

five columns, and an integer data type.

```
np.zeros((2, 5), dtype=int)
```

NumPy array.

```
np.arange(10)
```

`linspace` creates an array of evenly spaced floats given a start, stop
(inclusive), and number of elements.

```
np.linspace(0, 100, 5)
```

The `reshape` method changes an array into compatible dimensions.

```
np.arange(1, 8, 2).reshape(2, 2)
```

## Indexing and slicing arrays

Access elements by indexing (single element) or slicing (a series of
elements). Python is zero-indexed and stop points are not inclusive.

```
# *
product_array = np.array(["fruits", "vegetables", "cereal", "dairy", "eggs",
                          "snacks", "beverages", "coffee", "tea", "spices"])
```

Indexing a one-dimensional array. Index 1 grabs the second element; index
-1 grabs the last element (negative indexing works from the end).

```
product_array[1]
```

```
product_array[-1]
```

Slicing: with no start point it defaults to 0; with no step size it
defaults to 1.

```
product_array[:5]
```

```
product_array[5::2]
```

```
# *
product_array2D = product_array.reshape(2, 5)
```

Slicing a two-dimensional array requires coordinates: row then column.
Coffee is in the second row (index 1), third column (index 2).

```
product_array2D[1, 2]
```

Grab all rows, but slice columns from the third column to the end.

```
product_array2D[:, 2:]
```

Grab all columns, but only the second row onward.

```
product_array2D[1:, :]
```

## Array operations

Python's arithmetic operators can be used to perform array operations. With
NumPy there is no need to loop; the operation is applied element wise. The
concepts that make this work are vectorisation and broadcasting, which
eliminate the need for explicit loops in our code.

Add a scalar to every value in the array (for example, a tax or shipping
cost). NumPy broadcasts the 2 across every element.

```
# *
sales = [[0, 5, 155, 0, 518], [0, 1827, 616, 317, 325]]
sales_array = np.array(sales)
sales_array + 2
```

Multiply two arrays of the same length element wise. Here we slice the first
row into quantity and the second row into price, then multiply.

```
quantity = sales_array[0]
price = sales_array[1]
quantity * price
```

## Filtering arrays and modifying array values

We can filter arrays by indexing them with a logical test. Only the elements
at positions where the test returns True are returned in the resulting array.

A logical test on an array returns a Boolean array of True and False values.

```
# *
sales_array = np.array([0, 5, 155, 0, 518, 0, 1827, 616, 317, 325])
sales_array != 0
```

Passing the Boolean array back into the index filters out the values where the
test returned False (here, the zeros).

```
sales_array[sales_array != 0]
```

Combine logical tests with the pipe operator for or and the ampersand for and.
Unlike base Python, the words and and or cannot be used. Each condition must be
wrapped in parentheses.

```
sales_array[(sales_array == 616) | (sales_array < 100)]
```

An and condition mimics a between operator. Here we grab sales between 100 and
500, not inclusive.

```
sales_array[(sales_array > 100) & (sales_array < 500)]
```

Complex conditions can be stored in a variable, often called a mask, to keep
the filter readable.

```
mask = (sales_array > 100) & (sales_array < 500)
sales_array[mask]
```

We can filter one array by the values in another array of identical length.
Here we return products where sales were greater than zero.

```
# *
sales_array = np.array([0, 5, 155, 0, 518])
product_array = np.array(["fruits", "vegetables", "cereal", "dairy", "eggs"])
product_array[sales_array > 0]
```

Modifying array values. To assign a new value, specify the index and assign in
the new value.

```
# *
sales_array = np.array([0, 5, 155, 0, 518, 0, 1827, 616, 317, 325])
sales_array[1] = 25
sales_array
```

Filtering can also be used to change a set of values that meet a logical
criterion. Here any value equal to zero is changed to five.

```
sales_array[sales_array == 0] = 5
sales_array
```

## The where() function

The where function performs a logical test and returns one value where the test
is True and another where it is False. Syntax: `np.where(test, value_if_true,
value_if_false)`.

```
# *
inventory_array = np.array([10, 25, 5, 0, 0])
product_array = np.array(["fruits", "vegetables", "cereal", "dairy", "eggs"])
```

If inventory is less than or equal to zero, return "out of stock", otherwise
"in stock".

```
np.where(inventory_array <= 0, "out of stock", "in stock")
```

The false value can pull from another array. Here we return the product name
where stock exists, and "out of stock" otherwise.

```
np.where(inventory_array <= 0, "out of stock", product_array)
```

where can be chained. Even numbers stay "even"; odd numbers are checked again,

```
np.where(my_array % 2 == 0, "even", np.where(my_array == 9, my_array, "odd"))
```

## Challenge: Filtering and modifying arrays

Filter the product list to those with prices greater than 25. Then modify the
logic to force cola into the list, calling this array fancy_feast_special.
Finally, create a shipping cost array where shipping is zero if price is greater
than 20, otherwise five.

## Solution: Filtering and modifying arrays

```
# *
products = np.array(["salad", "bread", "mustard", "rare tomato", "cola",
                     "gourmet ice cream"])
prices = np.array([5.99, 6.99, 22.49, 99.99, 4.99, 49.99])
```

Filter products to those with a price greater than 25.

```
products[prices > 25]
```

Add a second condition with the or operator to force cola into the list, using
a mask to keep the filter readable.

```
mask = (prices > 25) | (products == "cola")
fancy_feast_special = products[mask]
fancy_feast_special
```

Create a shipping cost array: zero if price is greater than 20, otherwise five.

```
shipping = np.where(prices > 20, 0, 5)
shipping
```
