# Sum of triangles

You are given a function triangle that returns the triangular number,
for a given number, num. Your goal is to use this to create a function
called square that returns a the square of a given positive integer,
num.


## What are triangular numbers?

A triangular number is that number, plus all the numbers less than it.

For example:

> triangle(3) = 3 + 2 + 1 = 6
>
> triangle(4) = 4 + 3 + 2 + 1 = 10

## Squaring Up

**HINT 1:** Two triangles, when added together, make a square

**HINT 2:** Your two triangles will be of different sizes

**HINT 3:** Here are some examples for you:

triangle(2) + triangle(3) = 3 + 6 = 9

triangle(3) + triangle(4) = 6 + 10 = 16
